```python
missing = df.isnull().sum()
total_rows = len(df)
missing_percent = (missing / total_rows * 100).round(4)

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_percent
})

# Show only columns that have missing values
missing_df = missing_df[missing_df['Missing Count'] > 0]
print(missing_df.sort_values('Missing %', ascending=False))
```

```python
df=df.drop('End_Lat',axis=1)
print(df.columns)
```

```python
df.replace('',pd.NA, inplace=True)
df=df.dropna(subset=['Street','Zipcode','Timezone','City'])
print("Done:", df.shape)
```

```python
print(df.columns)
```

```python
df= df.drop('Airport_Code',axis =1)
print(df.columns)
```

```python
df.drop(['Wind_Chill(F)','Civil_Twilight','Nautical_Twilight','Astronomical_Twilight'],axis = 1, inplace = True)
```

```python
print(df[['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)']].head())
columns=['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)']
for col in columns:
    df[col]=df[col].fillna(df[col].median())
    print(f"Filled missing values in '{col}' with median: {df[col].median()}")
```

```python

print(df[['Wind_Direction', 'Weather_Condition', 'Sunrise_Sunset']].head())
cols = ['Wind_Direction', 'Weather_Condition', 'Sunrise_Sunset']

for col in cols:
    df[col] = df[col].fillna(df[col].mode()[0])
    print(f"Filled missing values in '{col}' with mode: {df[col].mode()[0]}")
   
```

```python
print(df.columns)
```

```python
df['Precipitation(in)']=df['Precipitation(in)'].fillna(0)
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='mixed', dayfirst=False)

# Now extract all useful time features
df['Hour']       = df['Start_Time'].dt.hour        # 0 to 23
df['Month']      = df['Start_Time'].dt.month       # 1 to 12
df['Year']       = df['Start_Time'].dt.year        # 2016 to 2023
df['DayOfWeek']  = df['Start_Time'].dt.day_name()  # Monday, Tuesday...
df['Season']     = df['Month'].map({
                    1:'Winter', 2:'Winter', 3:'Spring',
                    4:'Spring', 5:'Spring', 6:'Summer',
                    7:'Summer', 8:'Summer', 9:'Fall',
                    10:'Fall', 11:'Fall', 12:'Winter'})

print("New columns added!")
print(df[['Start_Time','Hour','Month','Year','DayOfWeek','Season']].head())
```

```python
# Calculate accident duration in minutes
df['Duration_Min'] = (pd.to_datetime(df['End_Time'], format='mixed') - 
                      df['Start_Time']).dt.total_seconds() / 60
```

```python
# Max realistic accident duration = 24 hours = 1440 minutes
df = df[df['Duration_Min'] <= 1440]
df = df[df['Duration_Min'] >= 1]

print("Duration cleaned!")
print(df['Duration_Min'].describe().round(2))
print("Final Shape:", df.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Makes charts look professional
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
```

```python
top_states = df['State'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_states.index, y=top_states.values, palette='Reds_r',hue=top_states.index,legend=False)
plt.title('Top 10 Most Dangerous States (2016-2023)', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_cities = df['City'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='Blues_r',hue=top_cities.index,legend=False)
plt.title('Top 10 Most Dangerous Cities (2016-2023)', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
day_night = df['Sunrise_Sunset'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(day_night.values, 
        labels=day_night.index, 
        autopct='%1.1f%%',
        colors=['#FFD700','#191970'],
        startangle=90,
        textprops={'fontsize':14})
plt.title('Day vs Night Accidents (2016-2023)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
hourly = df['Hour'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.lineplot(x=hourly.index, y=hourly.values, 
             color='red', linewidth=2.5, marker='o')
plt.title('Accidents by Hour of Day (2016-2023)', fontsize=16)
plt.xlabel('Hour (0=Midnight, 12=Noon)', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(range(0,24))
plt.tight_layout()
plt.show()
```

```python
# Set correct order of days
day_order = ['Monday','Tuesday','Wednesday',
             'Thursday','Friday','Saturday','Sunday']

day_counts = df['DayOfWeek'].value_counts().reindex(day_order)

plt.figure(figsize=(12,6))
sns.barplot(x=day_counts.index, y=day_counts.values, palette='viridis')
plt.title('Accidents by Day of Week (2016-2023)', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
severity = df['Severity'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=severity.index, y=severity.values, 
            palette='Reds_r', hue=severity.index, legend=False)
plt.title('Accidents by Severity Level (2016-2023)', fontsize=16)
plt.xlabel('Severity (1=Minor, 4=Critical)', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
season_order = ['Spring','Summer','Fall','Winter']
season_counts = df['Season'].value_counts().reindex(season_order)

plt.figure(figsize=(8,5))
sns.barplot(x=season_counts.index, y=season_counts.values,
            palette='coolwarm', hue=season_counts.index, legend=False)
plt.title('Accidents by Season (2016-2023)', fontsize=16)
plt.xlabel('Season', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.tight_layout()
plt.show()
```
